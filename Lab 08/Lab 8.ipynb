```python
def initial_centroid(df, k):
    return df.sample(k).reset_index(drop=True)
```

```python
def euclidean_distance(d, centroid):
    return np.sqrt(np.sum((d - centroid) ** 2))
```

```python
def assign_clusters(df, centroids):
    cluster_assignments = []
    for _, row in df.iterrows():
        distances = [euclidean_distance(row, centroids.iloc[i]) for i in range(len(centroids))]
        cluster_assignments.append(np.argmin(distances))
    return cluster_assignments
```

```python
def update_centroids(df, clusters, k):
    new_centroids = []
    for i in range(k):
        cluster_points = df[clusters == i]
        if not cluster_points.empty:
            new_centroids.append(cluster_points.mean())
        else:
            new_centroids.append(df.sample(1).iloc[0])
    return pd.DataFrame(new_centroids).reset_index(drop=True)
```

```python
def kmeans(df, k, max_iter=100):
    centroids = initial_centroid(df, k)

    for _ in range(max_iter):
        clusters = assign_clusters(df, centroids)
        df['cluster'] = clusters
        new_centroids = update_centroids(df.drop(columns='cluster'), df['cluster'], k)

        if centroids.equals(new_centroids):
            break

        centroids = new_centroids

    return df
```

```python
def silhouette_score(df, k):
    data = df.drop(columns='cluster').values
    labels = df['cluster'].values
    scores = []

    for i in range(len(data)):
        same_cluster = data[labels == labels[i]]
        other_clusters = [data[labels == label] for label in range(k) if label != labels[i]]

        if len(same_cluster) > 1:
            a = np.mean([np.linalg.norm(data[i] - point) for point in same_cluster if not np.array_equal(data[i], point)])
        else:
            a = 0

        b = np.inf
        for cluster in other_clusters:
            if len(cluster) > 0:
                dist = np.mean([np.linalg.norm(data[i] - point) for point in cluster])
                b = min(b, dist)

        s = (b - a) / max(a, b) if max(a, b) != 0 else 0
        scores.append(s)

    return np.mean(scores)
```

```python
result = kmeans(df, k=2)
score = silhouette_score(result, k=2)
print("Silhouette Score:", score)
```
